```python
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yfin
```

```python
start_date = '2017-01-01'
end_date = '2023-01-01'
```

```python
yfin.pdr_override()
# Fetch Apple Inc. data using the correct ticker symbol 'AAPL'
apple = yfin.download('AAPL', start=start_date, end=end_date)
```

```python
facebook = yfin.download('META', start=start_date, end=end_date)
```

```python
google = yfin.download('GOOGL', start=start_date, end=end_date)
```

```python
google.head()
```

```python
nvidia = yfin.download('NVDA', start=start_date, end=end_date)
```

```python
nvidia.head()
```

```python
tesla = yfin.download('TSLA', start=start_date, end=end_date)
```

```python
tesla.head()
```

```python
amazon = yfin.download('AMZN', start=start_date, end=end_date)
```

```python
amazon.head()
```

```python
dfs = [apple , facebook , google , nvidia , tesla , amazon]
```

```python
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day close price'] = df.Close.shift(1)
```

```python
apple.head()
```

```python
for df in dfs:
    df['Change in Price'] = df['Close'] - df['Previous day close price']
```

```python
apple.head()
```

```python
# (Close - Previous day close price )/previous day close price
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df["Previous day volume"] = df.Volume.shift(1)
```

```python
apple.head()
```

```python
for df in dfs :
    df["Change in volume"] = df['Volume'] - df['Previous day volume']
```

```python
apple.head()
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```
